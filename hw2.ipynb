### Определение: равномерное распределение
Абсолютно непрерывная случайная величина $X$ называется *равномерно распределённой* на отрезке $[a, b]$, если её плотность задаётся функцией 
$$p_{X}(x)=\frac{1}{b-a}\mathbb I_{[a, b]}(x),$$
где $\mathbb I_{[a, b]}$ — индикаторная функция отрезка $[a, b]$. Пишут:

$$\newcommand{\Uniform}{\mathop{\mathrm{Uniform}}}
X\sim \Uniform(a, b)$$

### Задача 1 (10 баллов)
Пусть $X \sim \Uniform(a, b)$, $b>a>0$, $Y=X^2$.

1. Найти плотность $p_Y(y)$. Является ли она ограниченной?
2. Пусть $y_1, \ldots, y_n$ — некоторая выборка, порождённая случайной величиной $Y$. Придумать какую-нибудь состоятельную оценку для параметра $a$ по этой выборке.
3. Является ли придуманная вами оценка несмещённой? (Подсказка: возьмите, например, $n=1$.)

__Решение__:

1. $p_Y(y) = \frac{1}{2 \sqrt{y} (b-a)} \mathbb I_{[a^2, b^2]}(y)$. Действительно, функция вероятности $F_Y(y) = \mathbb P_X([a, \sqrt{y}]) = \frac{\sqrt{y} - a}{b -a}$ для $y \in [a^2, b^2]$. Дифференцируем по $y$ и получаем результат. Данная функция является ограниченной, так как $0 < a < b$.

2. Рассмотрим функцию $\sqrt(min_n(y_n))$ (квадратный корень из минимума $n$ аргументов). Данная функция действительно является статистикой, так как прообраз любого луча $(r, \infty)$ есть множество вида $(x_1 > r^2, \dots, x_n > r^2)$, которое, очевидно, борелевское. Докажем состоятельность. Зафиксируем произвольную $\epsilon > 0$. Нам нужно рассмотреть меру события $\sqrt(min_n(y_n)) - a > \epsilon$, или, эквивалентно, $min_n(y_n) > (a + \epsilon)^2$. Последние возможно только если все $y_n$ больше $(a + \epsilon)^2$. По независимости выборки, вероятность этого события равна произведению $(1 - F_{y_n}((a^2 + \epsilon)^2))^n = (1 - \frac{\epsilon}{b - a})^n$. Последние означает, что мы возводим число в интервале $(0, 1)$ в степень $n$. Очевидно, что предел равен $0$, при $n \rightarrow \infty$. ЧТД.

3. Оценка выше не всегда является несмещённой. При $n = 1$ математическое ожидание нашей оценки будет равно $\int_{a^2}^{b^2} \frac{\sqrt{y}}{2 (b-a)} = \frac{a^2 + ab + b^2}{3}$. Например, если изначально нам было дано равномерное распределение на отрезке $[1, 100]$, значение выше не будет равно оцениваемому параметру.
